```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("CrimeGradeV3.csv")
df = df.drop_duplicates()
hp = pd.read_csv("HousePricesV3.csv")
hp = hp.drop('address', axis=1)
#hp['price'] = hp['price'].str.replace('[$,+,\s]', '')
hp['price'] = hp['price'].astype(int)
df.head()
grouped = hp.groupby('zipcode')['price'].mean().reset_index()
```

```python
df = df.reset_index(drop = True)
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax = sns.countplot(data=df, x='zipcode')

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

over = sns.barplot(x = 'zipcode', y = 'overall', data = df)

plt.xlabel("Zipcode", fontsize = 14)
plt.ylabel("Overall Crime Grade Score", fontsize = 14)
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax = sns.countplot(data=df, x='zipcode')

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

violent = sns.barplot(x = 'zipcode', y = 'violent', data = df)

plt.xlabel("Zipcode", fontsize = 14)
plt.ylabel("Violent Crime Grade Score", fontsize = 14)
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax = sns.countplot(data=df, x='zipcode')

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
property = sns.barplot(x = 'zipcode', y = 'property', data = df)

plt.xlabel("Zipcode", fontsize = 14)
plt.ylabel("Property Crime Grade Score", fontsize = 14)
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax = sns.countplot(data=df, x='zipcode')

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
other = sns.barplot(x = 'zipcode', y = 'other', data = df)

plt.xlabel("Zipcode", fontsize = 14)
plt.ylabel("Other Crime Grade Score", fontsize = 14)
```

```python
plt.figure(figsize=(12,8))
mergedData = grouped.merge(df, left_on = 'zipcode', right_on ='zipcode')
ax = sns.lineplot(data=mergedData, x = 'overall', y = 'price')
plt.xlabel("Overall Crime Stats", fontsize = 14)
plt.title("Distribution of Overall Crime", fontsize = 16);
```
